```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Load your dataset
df = pd.read_csv('nba_2022-23_stats.csv')

# Check for missing data (optional)

print(df.dtypes)
# We see that 3P object but it should be float
df['3P'] = pd.to_numeric(df['3P'], errors='coerce')
print(df.isnull().sum())
# Now all of the categories have correct datatypes
# But there are sum NaN values, with such a small dataset we dont want to lose any more entires. 
# We can find formulas for these attributes

def safe_divide(a, b):
    if pd.isna(a) or pd.isna(b):
        return np.nan
    if a == 0 and b == 0:
        return 0
    if b == 0:
        return np.nan
    return a / b

for index, row in df.iterrows():
    if pd.isna(row['FG%']):
        df.at[index, 'FG%'] = safe_divide(row['FG'], row['FGA'])
    
    if pd.isna(row['3P']):
        if row['3P%'] == 0 and row['3PA'] == 0:
            df.at[index, '3P'] = 0
        else:
            df.at[index, '3P'] = row['3P%'] * row['3PA']
    
    if pd.isna(row['3P%']):
        df.at[index, '3P%'] = safe_divide(row['3P'], row['3PA'])
    
    if pd.isna(row['2P%']):
        df.at[index, '2P%'] = safe_divide(row['2P'], row['2PA'])
    
    if pd.isna(row['eFG%']):
        df.at[index, 'eFG%'] = safe_divide(row['2P'] + 1.5 * row['3P'], row['FGA'])
    
    if pd.isna(row['FT%']):
        df.at[index, 'FT%'] = safe_divide(row['FT'], row['FTA'])
    
    if pd.isna(row['TS%']):
        df.at[index, 'TS%'] = safe_divide(0.5 * row['PTS'], row['FGA'] + 0.475 * row['FTA'])
    
    if pd.isna(row['3PAr']):
        df.at[index, '3PAr'] = safe_divide(row['3PA'], row['FGA'])
    
    if pd.isna(row['FTr']):
        df.at[index, 'FTr'] = safe_divide(row['FTA'], row['FGA'])
print(df.isnull().sum())
# There are still a lot of nan values, this is because the formulas of these include division, and when the denominator is 0
# then the value remains NaN.
# Well for percentages, if both numerator and denominator then result should be 0
print("Break")


nan_fg_rows = df[df['3P%'].isna()]

# Set pandas options to display all rows and columns
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns

print(nan_fg_rows)
```
